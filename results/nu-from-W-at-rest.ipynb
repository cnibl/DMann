# Neutrinos from $W$ bosons at rest
This notebook plots some results from simulations of $m_\chi=80.4$ GeV, i.e. producing the $W$ bosons at rest.


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Theta distribution

```python
cThPyt = np.genfromtxt("da-pyt8-mx80-ch24-int14-costh.dat", skip_header=7)
fig, ax = plt.subplots()
ax.plot(cThPyt[:,0],cThPyt[:,1],label=r"$\cos\theta$, Pythia")
ax.set_ylim(0,0.04)
ax.legend(loc="best")
```

## Energy distribution

```python
EPyt = np.genfromtxt("da-pyt8-mx80-ch24-int14.dat", skip_header=7)
fig, ax = plt.subplots()
ax.plot(EPyt[:,0],EPyt[:,1],label=r"$E_{\rm kin}$, Pythia")
ax.legend(loc="best")
```
